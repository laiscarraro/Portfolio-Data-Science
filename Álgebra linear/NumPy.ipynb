```python
a = np.matrix('1 2 4; 3 4 7')
b = np.matrix('8 1 6; 4 6 9')

def soma_matrizes(a,b):
  resp = []
  for linha in range(a.shape[0]):
    l = []
    for coluna in range(a.shape[1]):
      l.append(a[linha,coluna] + b[linha,coluna])
    resp.append(l)
  return np.matrix(resp)
```

```python
def teste_soma():
  tempo_final_np = []
  tempo_final_nosso = []
  for i in range(3):
    a = np.random.rand(10**(i+2), 10**(i+2))
    b = np.random.rand(10**(i+2), 10**(i+2))

    inicio_np = time.time()
    s = a + b
    tempo_final_np.append(time.time() - inicio_np)

    inicio_nosso = time.time()
    s = soma_matrizes(a,b)
    tempo_final_nosso.append(time.time() - inicio_nosso)

    del a
    del b

  return tempo_final_np, tempo_final_nosso
```

```python
import numpy as np
a = np.matrix('1 2; 3 4; 5 6')
b = np.matrix('1 2 3; 4 5 6')

def mult_matrizes(a,b):
  resp = []
  for i in range(a.shape[0]):
    l=[]
    for k in range(b.shape[1]): 
      teste = 0 
      for j in range(b.shape[0]):
        teste += b[j,k]*a[i,j] 
      l.append(teste)
    resp.append(l)
  return np.matrix(resp)
```

```python
def teste_mult():
  tempo_final_np = []
  tempo_final_nosso = []
  for i in range(3):
    a = np.random.rand(5**(i+1), 5**(i+1))
    b = np.random.rand(5**(i+1), 5**(i+1))

    inicio_np = time.time()
    s = a * b
    tempo_final_np.append(time.time() - inicio_np)

    inicio_nosso = time.time()
    s = mult_matrizes(a,b)
    tempo_final_nosso.append(time.time() - inicio_nosso)

    del a
    del b

  return tempo_final_np, tempo_final_nosso
```

```python
resultado = teste_mult()
print(resultado)
```
