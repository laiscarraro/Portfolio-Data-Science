## Conceitos

*Pergunta estatística* - Pergunta que pode ser respondida com dados.   
Ex.: **Quantas palavras em média um ser humano normal consegue ler por minuto?**

*População* - todos os indivíduos que queremos estudar  
*Variável de interesse* - característica da população que queremos estudar.  
Ex: A população são **todos os seres humanos normais**, a variável de interesse é a **quantidade de palavras que essa pessoa lê por minuto**.

*Amostra* - subconjunto de uma população.

E agora, como fazemos para coletar tudo isso?!

Muitas vezes, estudos estatísticos analisam uma população muito grande, em que seria inviável coletar dados de toda população. Ou ainda, mesmo que tivéssemos os dados de toda população, processá-los ainda seria um desafio. Por isso, uma ferramenta muito valiosa num estudo estatístico é a *amostragem*. Uma *amostra* é um subconjunto da população, geralmente de tamanho muito menor do que a população inteira. Os dados de alturas que coletamos mais cedo, por exemplo é uma amostra!!! Mas se ela é uma *boa* amostra, aí é outra história.....

### Como saber se uma amostra é boa?

Via de regra, uma amostra é boa se ela **representa bem a população**. Existem diversas técnicas de amostragem, que veremos a seguir, que têm como objetivo reduzir o *viés* de uma amostra, ou seja, tendências ou proporções de uma amostra que não representam as tendências de proporções de uma população. No entanto, mesmo assim ainda pode existir viés na coleta e/ou análise dos dados, dependendo de como é conduzido o estudo estatístico.

### 1. Probabilística vs não-probabilística

Tem a ver com a *probabilidade* de cada indivíduo ser escolhido para compor a amostra.

image.png

Levando em conta que a fórmula geral para calcular probabilidades é casos favoráveis sobre casos totais:

$P(A) = 1/5 = 20\%$

Fazendo o mesmo cálculo para cada uma das bolhinhas:

$P(B) = 1/5 = 20\%$  
$P(C) = 1/5 = 20\%$  
$P(D) = 1/5 = 20\%$  
$P(E) = 1/5 = 20\%$  

Logo, temos que:

$P(A) = P(B) = P(C) = P(D) = P(E) = 1/5 = 20\%$

Sendo assim, temos que num *sorteio qualquer* a probabilidade de cada um dos elementos de serem sorteados é **sempre igual**. Por isso, chamamos esse tipo de amostra de amostra **probabilística**.

````python
alturas = [1.65, 1.80, 1.58, 1.79, 1.60, 1.75, 1.70, 1.63, 1.81, 1.59, 1.61, 1.69, 1.73, 1.74, 1.62, 1.68, 1.72, 1.71, 1.57, 1.67]

import random
amostra = [random.choice(alturas) for i in range(10)]
amostra
````

````python
for i in range(20):
  print('Sorteio ' + str(i+1) + ': ' + str((1/len(alturas))*100) + '%') # temos 1 caso favorável entre 20 opções
````

Vamos supor agora que sou uma médica estudando o efeito de uma doença no sono das pessoas. No diagrama abaixo:

- Os **verdinhos** não possuem a doença
- Os **vermelhinhos** possuem a doença

Suponha que essa seja minha população inteira e minha amostra deve ser a maior possível.

image.png

$P(Maria) = 0\%$ (nunca escolheremos nenhum verdinho, pois não se encaixa nos objetivos do nosso estudo, então nunca escolheremos a Maria)

$P(José) = 100\%$ (para a maior amostra possível, com certeza escolheremos o José, pois escolheremos todos os vermelhinhos)

Logo, temos que:

$P(Maria) = 0\% \ne 100\% = P(José)$

Neste tipo de amostra, os indivíduos *não possuem* a mesma probabilidade de serem escolhidos, por isso são chamadas de *não probabilística*. Neste exemplo em específico, vimos a amostragem *intencional*, em que o/a pesquisador(a) escolhe os indivíduos que mais se encaixam em seu estudo.

### 2. Com reposição vs sem reposição

image.png

$P(azul)_{Sorteio1} = 1/5 = 20\%$  
$P(azul)_{Sorteio1} = P(vermelho)_{Sorteio1} = P(roxo)_{Sorteio1} = ... = 20\%$

Portanto, sabemos que a amostragem é probabilística. Além disso, temos que:

$P(azul)_{Sorteio2} = 1/5 = 20\% = P(azul)_{Sorteio1}$  

E isso vale para todo mundo:

$P(vermelho)_{Sorteio1} = P(vermelho)_{Sorteio2}$ e etc.

Portanto, numa amostra *com reposição*, temos que a probabilidade de os elementos serem escolhidos também é **igual entre os sorteios**.

````python
for i in range(20):
  print('Iteração ' + str(i+1) + ': ' + str((1/len(alturas))*100) + '%') # temos 1 caso favorável entre 20 opções
````

image.png

$P(azul)_{Sorteio1} = 1/5 = 20\%$  
$P(azul)_{Sorteio1} = P(vermelho)_{Sorteio1} = P(roxo)_{Sorteio1} = ... = 20\%$

Portanto, sabemos que a amostragem é probabilística. No entanto:

$P(azul)_{Sorteio2} = 1/4 = 25\% \ne 20\% = P(azul)_{Sorteio1}$

E isso também é verdade para os outros indivíduos: $P(verde)_{Sorteio1} \ne P(verde)_{Sorteio2}$ e etc. Mas, mesmo assim:

$P(azul)_{Sorteio2} = P(verde)_{Sorteio2} = ... = 25\%$

Ou seja, apesar de essa amostra não ter reposição, e os indivíduos tenham probabilidades de serem sorteados **diferentes entre os sorteios**, todos os indivíduos ainda têm a mesma probabilidade de serem sorteados *num mesmo sorteio*.

````python
amostra = set()

while len(amostra) < 5:
  sorteado = random.choice(alturas)
  amostra.add(sorteado)

list(amostra)
````

````python
retirados = set()

while len(retirados) < len(alturas) - 1: # sortear até só sobrar 1 elemento
  tam_antes = len(retirados)

  retirados.add(random.randint(0,len(alturas)-1))

  tam_depois = len(retirados)

  if tam_antes < tam_depois:
    print('Iteração ' + str(len(retirados)+1) + ': ' + str(round((1/(len(alturas)-len(retirados)))*100, 2)) + '%') # favoráveis: total - retirados
````

#### Outras possíveis soluções

````python
# Solução utilizando uma lista de índices

amostra = []
indices = []

while len(amostra) < 8:
  sorteado = random.randint(0,len(alturas)-1)
  if sorteado not in indices:
    amostra.append(alturas[sorteado])
    indices.append(sorteado)

amostra
````

````python
# Solução retirando os ítens do conjunto de dados

populacao = alturas.copy()
amostra = []

while len(amostra) < 8:
  sorteado = random.randint(0,len(populacao)-1)
  amostra.append(populacao.pop(sorteado))

amostra
````

````python
# Solução utilizando conjuntos

retirados = set()

while len(retirados) < 5:
  retirados.add(alturas[random.randint(0,len(alturas)-1)])

print(list(retirados))
````

````python
# Solução utilizando conjuntos levando em conta valores repetidos

amostra = set([])

while len(amostra) < 8:
  sorteado = random.randint(0, len(alturas) - 1)
  amostra.add((sorteado, alturas[sorteado]))

amostra = [j for i,j in list(amostra)]
amostra
````

Sabemos que a altura de alunos do maternal é significativamente diferente do que a de alunos do ensino médio. Sabendo que alunos do maternal são uma minoria, se quisermos fazer uma amostra *representativa*, devemos garantir que as proporções de nível de ensino sejam mantidas na amostra. Por isso, podemos usar a amostra *estratificada*.

image.png

````python
# filtrando os alunos
alunos_maternal = [aluno for aluno in alunos if aluno[1] == 'Maternal']
alunos_fundamental = [aluno for aluno in alunos if aluno[1] == 'Fundamental']
alunos_medio = [aluno for aluno in alunos if aluno[1] == 'Médio']

# calculando proporcoes
p_maternal = round(100*(len(alunos_maternal)/len(alunos)))
p_fundamental = round(100*(len(alunos_fundamental)/len(alunos)))
p_medio = round(100*(len(alunos_medio)/len(alunos)))

print('Proporção maternal: ', p_maternal, '%', sep='')
print('Proporção fundamental: ', p_fundamental, '%', sep='')
print('Proporção médio: ', p_medio, '%', sep='')
````

````python
# pegando uma amostra de 10%
tam_amostra = round((10*len(alunos))/100)

tam_maternal = round(tam_amostra*(p_maternal/100))
tam_fundamental = round(tam_amostra*(p_fundamental/100))
tam_medio = round(tam_amostra*(p_medio/100))

# pegando só as alturas aleatórias
maternal = [alunos_maternal[random.randint(0, len(alunos_maternal) - 1)][0] for i in range(tam_maternal)]
fundamental = [alunos_fundamental[random.randint(0, len(alunos_fundamental) - 1)][0] for i in range(tam_fundamental)]
medio = [alunos_medio[random.randint(0, len(alunos_medio) - 1)][0] for i in range(tam_medio)]

# juntando tudo
estratificada = maternal + fundamental + medio

print(estratificada)
print('Média real:', sum([i[0] for i in alunos])/len(alunos))
print('Estimativa:', sum(estratificada)/len(estratificada))
````

Às vezes, é muito custoso aplicar na prática a amostragem estratificada, ou até mesmo a aleatória simples. Por isso, podemos simplificá-la com a abordagem de conglomerados. No nosso exemplo, cada sala é um conglomerado, e sorteamos *as salas*.
