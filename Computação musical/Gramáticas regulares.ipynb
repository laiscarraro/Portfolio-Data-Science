# Gramáticas regulares

Vamos construir uma gramática para reconhecer as escalas maiores e menores \o/

```python
from music21 import *

ps_raw = ['C', 'D', 'E', 'F', 'G', 'A', 'B']
ps = ps_raw + [i+'b' for i in ps_raw] + [i+'#' for i in ps_raw]

exemplos = [[n.name for n in scale.MajorScale(p).pitches] for p in ps] #+ [[n.name for n in scale.MelodicMinorScale(p).pitches] for p in ps]

#exemplos = [[n.name for n in scale.MajorScale('C').pitches], [n.name for n in scale.MelodicMinorScale('C').pitches]]
```

```python
def gramatica(exemplos):
  alfabeto = list(set([item for sublist in exemplos for item in sublist]))
  estados = []
  producoes = []

  for i in range(len(exemplos[0])):
    estados.append('s'+str(len(estados)))
    transicoes = []
    for e in exemplos:
      transicoes.append(e[i])
    for t in list(set(transicoes)):
      right = ''
      if i < len(exemplos[0]) - 1: 
        right = ' s'+str(i+1)
      producoes.append([estados[-1], t+right])
  return producoes

def reconhecer(cadeia, g):
  indice = 0
  caminho = []
  reconhece = True
  for c in cadeia:
    try:
      regras = [regra for regra in g if regra[0][1] == str(indice)]
      len_antigo = len(caminho)
      for r in regras:
        if c == r[1].split(' ')[0]:
          caminho.append(r)
      if len(caminho) == len_antigo:
        caminho.append(['Produção inexistente (' + regras[0][0] + ' -> ' + c + ')', 'NÃO RECONHECE'])
        reconhece = False
        break
      indice += 1
    except:
      caminho.append(['Estado inexistente (s' + str(indice) + ')', 'NÃO RECONHECE'])
      reconhece = False
      break

  if reconhece:
    if len(caminho[-1][1].split(' ')) > 1:
      caminho.append(['Estado não terminal (' + caminho[-1][1] + ')', 'NÃO RECONHECE'])
      reconhece = False
    else:
      caminho[-1][1] += '\nRECONHECE'
  
  return caminho, reconhece
```

```python
g = gramatica(exemplos)
cadeia = [n.name for n in scale.MajorScale('C-').pitches]

caminho, reconhece = reconhecer(cadeia, g)
for c in caminho:
  print(c[0], '->', c[1])
```
